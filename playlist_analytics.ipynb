```python
def get_info(url: str) -> dict:
    """Fetches about a YouTube video.

    Note:
        Returns 1.title, 2.views, 3.date, 4.likes

    args:
        url (str): The URL of YouTube video.

    Returns:
        dict: A dictionary containing the videos title, ...
    """ 
    yt = YouTube(url)

    data = {
        "title" : yt.title,
        "views" : yt.views,
        "data" : yt.publish_date,
        "likes" : yt.likes,
        "length" : yt.length
    }
    return data
```
