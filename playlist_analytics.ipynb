```python
import time
from tqdm import tqdm  
from pytubefix import Playlist, YouTube
```

```python
URL = "https://www.youtube.com/watch?v=aloH6wQKMuw&list=PLDDuEZQEC_uALLtKyjvhf0L9uhaibZ6Vn"
```

```python
def get_info(url: str) -> dict:
    """Fetches basic info about a YouTube video.

    Note:
        Returns 1.title, 2.views, 3.date, 4.likes

    args:
        url (str): The URL of YouTube video.

    Returns:
        dict: A dictionary containing the videos title, views, date, likes.
    """ 
    yt = YouTube(url)

    data = {
        "title" : yt.title,
        "views" : yt.views,
        "data" : yt.publish_date.strftime("%Y-%m-%d %H:%M:%S"),
        "likes" : yt.likes,
        "length" : yt.length
    }
    return data
```

```python
URL = "https://www.youtube.com/watch?v=aloH6wQKMuw&list=PLDDuEZQEC_uALLtKyjvhf0L9uhaibZ6Vn"

all_videos = []

pl = Playlist(URL)
```

```python
for i in tqdm(pl.videos):
    all_videos.append(get_info(i.watch_url))
all_videos.sort(key = lambda x: x['views'])
```
